```python
def ensure_clean_dir(path):
    path = Path(path)
    if path.exists():
        shutil.rmtree(path)
    path.mkdir(parents=True, exist_ok=True)
    return path


def list_image_files(image_dir):
    image_dir = Path(image_dir)
    patterns = ('*.jpg', '*.jpeg', '*.png', '*.bmp', '*.tif', '*.tiff', '*.webp')
    files = []
    for pattern in patterns:
        files.extend(sorted(image_dir.glob(pattern)))
    return sorted(set(files))


def split_dirs(dataset_root, split_name):
    dataset_root = Path(dataset_root)
    return dataset_root / split_name / 'images', dataset_root / split_name / 'labels'


def looks_like_yolo_dataset(dataset_root):
    dataset_root = Path(dataset_root)
    for split_name in ('train', 'val', 'test'):
        image_dir, label_dir = split_dirs(dataset_root, split_name)
        if image_dir.exists() and label_dir.exists():
            return True
    return False


def stable_fraction(key):
    digest = hashlib.md5(str(key).encode('utf-8')).hexdigest()
    return int(digest[:8], 16) / 0xFFFFFFFF


def download_file(url, output_path):
    output_path = Path(output_path)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    if output_path.exists():
        print('already exists:', output_path)
        return output_path
    print('downloading', url, '->', output_path)
    urllib.request.urlretrieve(url, output_path)
    return output_path


def extract_archive(archive_path, output_dir):
    archive_path = Path(archive_path)
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    if archive_path.suffix == '.zip':
        with zipfile.ZipFile(archive_path, 'r') as zf:
            zf.extractall(output_dir)
    elif archive_path.suffix in {'.gz', '.tgz'} or archive_path.name.endswith('.tar.gz'):
        with tarfile.open(archive_path, 'r:*') as tf:
            tf.extractall(output_dir)
    else:
        raise ValueError(f'Unsupported archive type: {archive_path}')


def prepare_coco_files():
    COCO_ROOT.mkdir(parents=True, exist_ok=True)
    archive_dir = COCO_ROOT / 'archives'
    archive_dir.mkdir(parents=True, exist_ok=True)

    for filename, url in COCO_DOWNLOADS.items():
        archive_path = download_file(url, archive_dir / filename)
        extract_archive(archive_path, COCO_ROOT)


def prepare_archive_source(target_root, local_archive_path=None, download_url=''):
    target_root = Path(target_root)
    if any(target_root.iterdir()) if target_root.exists() else False:
        print('using existing extracted source:', target_root)
        return True

    if local_archive_path:
        local_archive_path = Path(local_archive_path)
        if local_archive_path.exists():
            print('extracting local archive:', local_archive_path)
            extract_archive(local_archive_path, target_root)
            return True

    if download_url:
        archive_name = download_url.rstrip('/').split('/')[-1] or 'dataset.zip'
        archive_path = download_file(download_url, target_root.parent / archive_name)
        extract_archive(archive_path, target_root)
        return True

    print('source missing for', target_root)
    return False


def write_yolo_labels(label_path, rows):
    label_path = Path(label_path)
    label_path.parent.mkdir(parents=True, exist_ok=True)
    with open(label_path, 'w', encoding='utf-8') as f:
        for class_id, x_center, y_center, width, height in rows:
            f.write(f'{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n')


def write_dataset_yaml(dataset_root, yaml_path, class_names):
    dataset_root = Path(dataset_root)
    yaml_path = Path(yaml_path)
    data = {
        'path': str(dataset_root),
        'train': 'train/images',
        'val': 'val/images',
        'names': {idx: name for idx, name in enumerate(class_names)},
    }
    if (dataset_root / 'test' / 'images').exists():
        data['test'] = 'test/images'
    with open(yaml_path, 'w', encoding='utf-8') as f:
        yaml.safe_dump(data, f, sort_keys=False, allow_unicode=True)


def apply_vignette(gray_image, strength=0.08):
    if strength <= 0:
        return gray_image

    h, w = gray_image.shape[:2]
    y = np.linspace(-1.0, 1.0, h, dtype=np.float32)
    x = np.linspace(-1.0, 1.0, w, dtype=np.float32)
    xx, yy = np.meshgrid(x, y)
    radius = np.sqrt(xx ** 2 + yy ** 2)
    vignette = 1.0 - strength * np.clip(radius, 0.0, 1.0)
    return np.clip(gray_image.astype(np.float32) * vignette, 0.0, 255.0).astype(np.uint8)


def white_hot_bgr(image_bgr, cfg):
    gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    if cfg['apply_clahe']:
        clahe = cv2.createCLAHE(
            clipLimit=cfg['clahe_clip_limit'],
            tileGridSize=(cfg['clahe_tile_size'], cfg['clahe_tile_size'])
        )
        gray = clahe.apply(gray)

    gray_f = gray.astype(np.float32) / 255.0
    blur_small = cv2.GaussianBlur(gray_f, (0, 0), 1.2)
    detail = np.clip(gray_f - blur_small, -1.0, 1.0)
    heat = gray_f + cfg['hotspot_gain'] * np.maximum(detail, 0.0) + cfg['edge_gain'] * np.abs(detail)
    heat = np.clip(heat, 0.0, 1.0)
    heat = np.power(heat, cfg['gamma'])
    heat = (heat * 255.0).astype(np.uint8)
    heat = apply_vignette(heat, cfg['vignette_strength'])

    if cfg['downsample_scale'] < 1.0:
        h, w = heat.shape[:2]
        small_w = max(1, int(w * cfg['downsample_scale']))
        small_h = max(1, int(h * cfg['downsample_scale']))
        heat = cv2.resize(heat, (small_w, small_h), interpolation=cv2.INTER_AREA)
        heat = cv2.resize(heat, (w, h), interpolation=cv2.INTER_LINEAR)

    if cfg['blur_sigma'] > 0:
        heat = cv2.GaussianBlur(heat, (0, 0), cfg['blur_sigma'])

    if cfg['sensor_noise_std'] > 0:
        noise = np.random.normal(0.0, cfg['sensor_noise_std'], heat.shape).astype(np.float32)
        heat = np.clip(heat.astype(np.float32) + noise, 0.0, 255.0).astype(np.uint8)

    return cv2.cvtColor(heat, cv2.COLOR_GRAY2BGR)


def summarize_dataset(dataset_root):
    dataset_root = Path(dataset_root)
    summary = {}
    for split_name in ('train', 'val', 'test'):
        image_dir, label_dir = split_dirs(dataset_root, split_name)
        if not image_dir.exists() or not label_dir.exists():
            continue
        image_files = list_image_files(image_dir)
        label_files = sorted(label_dir.glob('*.txt'))
        class_counter = Counter()
        for label_path in label_files:
            for line in label_path.read_text(encoding='utf-8').splitlines():
                line = line.strip()
                if not line:
                    continue
                class_id = int(float(line.split()[0]))
                class_counter[class_id] += 1
        summary[split_name] = {
            'images': len(image_files),
            'labels': len(label_files),
            'boxes_per_class': {CLASS_NAMES[k]: v for k, v in sorted(class_counter.items())},
        }
    return summary


def print_dataset_summary(dataset_root, title=None):
    if title:
        print('\n' + title)
    summary = summarize_dataset(dataset_root)
    if not summary:
        print('no dataset found at', dataset_root)
        return
    for split_name, split_summary in summary.items():
        print(split_name, split_summary)


def thermalize_dataset(source_root, target_root, cfg, convert_train=True, convert_val=True, convert_test=True):
    source_root = Path(source_root)
    target_root = ensure_clean_dir(target_root)
    convert_lookup = {
        'train': convert_train,
        'val': convert_val,
        'test': convert_test,
    }

    for split_name in ('train', 'val', 'test'):
        source_image_dir, source_label_dir = split_dirs(source_root, split_name)
        if not source_image_dir.exists():
            continue

        target_image_dir, target_label_dir = split_dirs(target_root, split_name)
        target_image_dir.mkdir(parents=True, exist_ok=True)
        target_label_dir.mkdir(parents=True, exist_ok=True)

        for label_path in sorted(source_label_dir.glob('*.txt')):
            shutil.copy2(label_path, target_label_dir / label_path.name)

        for image_path in list_image_files(source_image_dir):
            image = cv2.imread(str(image_path), cv2.IMREAD_COLOR)
            if image is None:
                print('skip unreadable image:', image_path)
                continue
            output_image = white_hot_bgr(image, cfg) if convert_lookup[split_name] else image
            ok = cv2.imwrite(str(target_image_dir / image_path.name), output_image)
            if not ok:
                raise RuntimeError(f'Failed to write image: {target_image_dir / image_path.name}')

    return target_root


def path_size_mb(path):
    path = Path(path)
    if not path.exists():
        return 0.0
    if path.is_file():
        return path.stat().st_size / (1024 * 1024)
    total = 0
    for child in path.rglob('*'):
        if child.is_file():
            total += child.stat().st_size
    return total / (1024 * 1024)


def cleanup_paths(paths_to_remove, label='cleanup'):
    print(f'\n{label}: start')
    for path in paths_to_remove:
        path = Path(path)
        if not path.exists():
            continue
        size_mb = path_size_mb(path)
        if path.is_dir():
            shutil.rmtree(path)
        else:
            path.unlink()
        print(f'removed {path} ({size_mb:.1f} MB)')
    print(f'{label}: done')


def mount_drive_for_backup_if_needed():
    from google.colab import drive
    drive_root = Path('/content/drive')
    if not drive_root.exists() or not any(drive_root.iterdir()):
        drive.mount('/content/drive')


def copy_dir_replace(src, dst):
    src = Path(src)
    dst = Path(dst)
    if dst.exists():
        shutil.rmtree(dst)
    shutil.copytree(src, dst)


def zip_directory(source_dir, zip_path):
    source_dir = Path(source_dir)
    zip_path = Path(zip_path)
    zip_path.parent.mkdir(parents=True, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'w', compression=zipfile.ZIP_DEFLATED) as zip_file:
        for child in source_dir.rglob('*'):
            if child.is_file():
                zip_file.write(child, arcname=str(child.relative_to(source_dir.parent)))


def extract_zip_to_parent(zip_path, target_dir):
    zip_path = Path(zip_path)
    target_dir = Path(target_dir)
    target_dir.parent.mkdir(parents=True, exist_ok=True)
    if target_dir.exists():
        shutil.rmtree(target_dir)
    with zipfile.ZipFile(zip_path, 'r') as zip_file:
        zip_file.extractall(target_dir.parent)


def backup_thermal_dataset_to_drive(local_dataset_root, drive_dataset_root, drive_zip_path=None):
    local_dataset_root = Path(local_dataset_root)
    drive_dataset_root = Path(drive_dataset_root)
    mount_drive_for_backup_if_needed()
    drive_dataset_root.parent.mkdir(parents=True, exist_ok=True)
    copy_dir_replace(local_dataset_root, drive_dataset_root)
    print(f'Backed up thermal dataset folder -> {drive_dataset_root} ({path_size_mb(drive_dataset_root):.1f} MB)')
    if drive_zip_path is not None:
        zip_directory(local_dataset_root, drive_zip_path)
        print(f'Backed up thermal dataset zip -> {drive_zip_path} ({path_size_mb(drive_zip_path):.1f} MB)')


def restore_thermal_dataset_from_drive(local_dataset_root, drive_dataset_root, drive_zip_path=None):
    local_dataset_root = Path(local_dataset_root)
    drive_dataset_root = Path(drive_dataset_root)
    if local_dataset_root.exists() and validate_thermal_dataset(local_dataset_root):
        return True
    mount_drive_for_backup_if_needed()
    if drive_dataset_root.exists():
        copy_dir_replace(drive_dataset_root, local_dataset_root)
        print(f'Restored thermal dataset folder from Drive -> {local_dataset_root}')
        return validate_thermal_dataset(local_dataset_root)
    if drive_zip_path is not None and Path(drive_zip_path).exists():
        extract_zip_to_parent(drive_zip_path, local_dataset_root)
        print(f'Restored thermal dataset zip from Drive -> {local_dataset_root}')
        return validate_thermal_dataset(local_dataset_root)
    return False


def validate_thermal_dataset(dataset_root):
    dataset_root = Path(dataset_root)
    for split_name in ('train', 'val', 'test'):
        image_dir, label_dir = split_dirs(dataset_root, split_name)
        if image_dir.exists() and label_dir.exists() and list_image_files(image_dir):
            return True
    return False
```

```python
def normalize_existing_yolo_dataset(source_root, target_root, source_class_to_target_class, prefix):
    source_root = Path(source_root)
    target_root = ensure_clean_dir(target_root)

    if not looks_like_yolo_dataset(source_root):
        print('skip non-yolo source:', source_root)
        return False

    for split_name in ('train', 'val', 'test'):
        source_image_dir, source_label_dir = split_dirs(source_root, split_name)
        if not source_image_dir.exists() or not source_label_dir.exists():
            continue

        target_image_dir, target_label_dir = split_dirs(target_root, split_name)
        target_image_dir.mkdir(parents=True, exist_ok=True)
        target_label_dir.mkdir(parents=True, exist_ok=True)

        for image_path in list_image_files(source_image_dir):
            target_name = f'{prefix}_{image_path.name}'
            label_path = source_label_dir / f'{image_path.stem}.txt'
            target_image_path = target_image_dir / target_name
            target_label_path = target_label_dir / f'{Path(target_name).stem}.txt'
            shutil.copy2(image_path, target_image_path)

            rows = []
            if label_path.exists():
                for line in label_path.read_text(encoding='utf-8').splitlines():
                    line = line.strip()
                    if not line:
                        continue
                    parts = line.split()
                    class_id = int(float(parts[0]))
                    if class_id not in source_class_to_target_class:
                        continue
                    rows.append((source_class_to_target_class[class_id], *map(float, parts[1:5])))
            write_yolo_labels(target_label_path, rows)

    return True


def build_coco_subset(coco_root, target_root, class_names):
    coco_root = Path(coco_root)
    target_root = ensure_clean_dir(target_root)
    target_class_set = set(class_names)
    split_mapping = {'train2017': 'train', 'val2017': 'val'}

    for coco_split, output_split in split_mapping.items():
        ann_path = coco_root / 'annotations' / f'instances_{coco_split}.json'
        image_dir = coco_root / coco_split
        if not ann_path.exists() or not image_dir.exists():
            print('skip missing coco split:', coco_split)
            continue

        with open(ann_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        categories = {item['id']: item['name'] for item in data['categories']}
        image_info = {item['id']: item for item in data['images']}
        annotations_by_image = defaultdict(list)

        for ann in data['annotations']:
            category_name = categories.get(ann['category_id'])
            if category_name not in target_class_set:
                continue
            annotations_by_image[ann['image_id']].append(ann)

        target_image_dir, target_label_dir = split_dirs(target_root, output_split)
        target_image_dir.mkdir(parents=True, exist_ok=True)
        target_label_dir.mkdir(parents=True, exist_ok=True)

        copied = 0
        for image_id, anns in annotations_by_image.items():
            info = image_info[image_id]
            src_image_path = image_dir / info['file_name']
            if not src_image_path.exists():
                continue

            target_name = f'coco_{coco_split}_{info["file_name"]}'
            shutil.copy2(src_image_path, target_image_dir / target_name)

            rows = []
            width = float(info['width'])
            height = float(info['height'])
            for ann in anns:
                x, y, w, h = ann['bbox']
                x_center = (x + w / 2.0) / width
                y_center = (y + h / 2.0) / height
                box_w = w / width
                box_h = h / height
                class_name = categories[ann['category_id']]
                rows.append((CLASS_TO_ID[class_name], x_center, y_center, box_w, box_h))
            write_yolo_labels(target_label_dir / f'{Path(target_name).stem}.txt', rows)
            copied += 1

        print(f'coco {coco_split}: copied {copied} images')

    return True


def parse_antiuav_annotation_file(annotation_path):
    annotation_path = Path(annotation_path)
    if annotation_path.suffix.lower() == '.json':
        payload = json.loads(annotation_path.read_text(encoding='utf-8'))
        if isinstance(payload, list):
            return payload
        if isinstance(payload, dict):
            for key in ('annotations', 'boxes', 'gt_rect', 'bboxes'):
                if key in payload and isinstance(payload[key], list):
                    return payload[key]
        return []

    rows = []
    for line in annotation_path.read_text(encoding='utf-8').splitlines():
        line = line.strip().replace(',', ' ')
        if not line:
            continue
        parts = [p for p in line.split() if p]
        if len(parts) < 4:
            continue
        rows.append([float(parts[0]), float(parts[1]), float(parts[2]), float(parts[3])])
    return rows


def normalize_antiuav_dataset(source_root, target_root, prefix='antiuav'):
    source_root = Path(source_root)
    if looks_like_yolo_dataset(source_root):
        return normalize_existing_yolo_dataset(source_root, target_root, {0: CLASS_TO_ID['drone']}, prefix)

    target_root = ensure_clean_dir(target_root)
    image_dir_candidates = ('images', 'img', 'IR', 'infrared', 'frames')
    ann_name_candidates = ('gt_rect.txt', 'groundtruth.txt', 'gt.txt', 'IR_label.json', 'labels.json', 'annotations.json')
    found_sequences = 0

    for sequence_dir in sorted([p for p in source_root.iterdir() if p.is_dir()]):
        image_dir = None
        for candidate in image_dir_candidates:
            matches = list(sequence_dir.rglob(candidate))
            if matches:
                image_dir = matches[0]
                break
        if image_dir is None:
            continue

        ann_path = None
        for candidate in ann_name_candidates:
            matches = list(sequence_dir.rglob(candidate))
            if matches:
                ann_path = matches[0]
                break
        if ann_path is None:
            continue

        image_files = list_image_files(image_dir)
        annotations = parse_antiuav_annotation_file(ann_path)
        if not image_files or not annotations:
            continue

        found_sequences += 1
        split_name = 'train' if stable_fraction(sequence_dir.name) < 0.8 else 'val'
        target_image_dir, target_label_dir = split_dirs(target_root, split_name)
        target_image_dir.mkdir(parents=True, exist_ok=True)
        target_label_dir.mkdir(parents=True, exist_ok=True)

        for image_path, ann in zip(image_files, annotations):
            image = cv2.imread(str(image_path), cv2.IMREAD_COLOR)
            if image is None:
                continue
            h, w = image.shape[:2]
            if isinstance(ann, dict):
                bbox = ann.get('bbox') or ann.get('rect') or ann.get('box') or []
            else:
                bbox = ann
            if len(bbox) < 4:
                continue
            x, y, bw, bh = map(float, bbox[:4])
            if bw <= 1 or bh <= 1:
                continue
            x_center = (x + bw / 2.0) / w
            y_center = (y + bh / 2.0) / h
            box_w = bw / w
            box_h = bh / h
            target_name = f'{prefix}_{sequence_dir.name}_{image_path.name}'
            shutil.copy2(image_path, target_image_dir / target_name)
            write_yolo_labels(
                target_label_dir / f'{Path(target_name).stem}.txt',
                [(CLASS_TO_ID['drone'], x_center, y_center, box_w, box_h)],
            )

    if found_sequences == 0:
        print('Anti-UAV source was not converted. Provide a YOLO-style archive or a sequence archive with frame folders and bbox files.')
        return False
    return True


def merge_normalized_datasets(source_roots, merged_root):
    merged_root = ensure_clean_dir(merged_root)

    for source_root in source_roots:
        source_root = Path(source_root)
        if not source_root.exists():
            continue
        for split_name in ('train', 'val', 'test'):
            source_image_dir, source_label_dir = split_dirs(source_root, split_name)
            if not source_image_dir.exists() or not source_label_dir.exists():
                continue
            target_image_dir, target_label_dir = split_dirs(merged_root, split_name)
            target_image_dir.mkdir(parents=True, exist_ok=True)
            target_label_dir.mkdir(parents=True, exist_ok=True)
            for image_path in list_image_files(source_image_dir):
                shutil.copy2(image_path, target_image_dir / image_path.name)
                label_path = source_label_dir / f'{image_path.stem}.txt'
                if label_path.exists():
                    shutil.copy2(label_path, target_label_dir / label_path.name)
                else:
                    write_yolo_labels(target_label_dir / f'{image_path.stem}.txt', [])

    return merged_root


def ensure_test_split(dataset_root, test_fraction=0.5):
    dataset_root = Path(dataset_root)
    test_image_dir, test_label_dir = split_dirs(dataset_root, 'test')
    if test_image_dir.exists() and list_image_files(test_image_dir):
        return

    val_image_dir, val_label_dir = split_dirs(dataset_root, 'val')
    if not val_image_dir.exists():
        return

    test_image_dir.mkdir(parents=True, exist_ok=True)
    test_label_dir.mkdir(parents=True, exist_ok=True)

    for image_path in list_image_files(val_image_dir):
        if stable_fraction(image_path.name) < test_fraction:
            label_path = val_label_dir / f'{image_path.stem}.txt'
            shutil.move(str(image_path), str(test_image_dir / image_path.name))
            if label_path.exists():
                shutil.move(str(label_path), str(test_label_dir / label_path.name))
```

```python
import time
from google.colab import drive

TRAIN_DATA_YAML = str(THERMAL_DATA_YAML)
TRAIN_BATCH = 64
TRAIN_TOTAL_EPOCHS = 200
TRAIN_CHUNK_EPOCHS = 50
TRAIN_IMGSZ = 640
RUN_NAME = 'multiclass_whitehot_chunked'
RUNS_DETECT_ROOT = CONTENT_ROOT / 'runs' / 'detect'
LOCAL_RUN_DIR = RUNS_DETECT_ROOT / RUN_NAME
DRIVE_BACKUP_ROOT = Path('/content/drive/MyDrive/1.non_modelfile_train/1.Train_Final/non_test/Final/TrainAndTest_model/YOLO8m_RGB/multiclass_whitehot_chunked_backup')
DRIVE_RUN_DIR = DRIVE_BACKUP_ROOT / 'latest_run'
DRIVE_STATE_PATH = DRIVE_BACKUP_ROOT / 'train_state.json'


def mount_drive_if_needed():
    drive_root = Path('/content/drive')
    if not drive_root.exists() or not any(drive_root.iterdir()):
        drive.mount('/content/drive')


def copytree_replace(src, dst):
    src = Path(src)
    dst = Path(dst)
    if dst.exists():
        shutil.rmtree(dst)
    shutil.copytree(src, dst)


def read_drive_state():
    if DRIVE_STATE_PATH.exists():
        return json.loads(DRIVE_STATE_PATH.read_text(encoding='utf-8'))
    return {'completed_epochs': 0}


def write_drive_state(payload):
    DRIVE_BACKUP_ROOT.mkdir(parents=True, exist_ok=True)
    DRIVE_STATE_PATH.write_text(json.dumps(payload, indent=2), encoding='utf-8')


def restore_from_drive_if_needed():
    state = read_drive_state()
    if LOCAL_RUN_DIR.exists():
        return state
    if DRIVE_RUN_DIR.exists():
        LOCAL_RUN_DIR.parent.mkdir(parents=True, exist_ok=True)
        copytree_replace(DRIVE_RUN_DIR, LOCAL_RUN_DIR)
        print('Restored local run from Google Drive backup:', DRIVE_RUN_DIR)
    return state


def backup_run_to_drive(completed_epochs):
    if not LOCAL_RUN_DIR.exists():
        raise RuntimeError(f'Local run directory missing: {LOCAL_RUN_DIR}')
    DRIVE_BACKUP_ROOT.mkdir(parents=True, exist_ok=True)
    copytree_replace(LOCAL_RUN_DIR, DRIVE_RUN_DIR)

    last_checkpoint = LOCAL_RUN_DIR / 'weights' / 'last.pt'
    best_checkpoint = LOCAL_RUN_DIR / 'weights' / 'best.pt'
    if last_checkpoint.exists():
        shutil.copy2(last_checkpoint, DRIVE_BACKUP_ROOT / 'last.pt')
    if best_checkpoint.exists():
        shutil.copy2(best_checkpoint, DRIVE_BACKUP_ROOT / 'best.pt')

    results_csv = LOCAL_RUN_DIR / 'results.csv'
    if results_csv.exists():
        shutil.copy2(results_csv, DRIVE_BACKUP_ROOT / 'results.csv')

    payload = {
        'run_name': RUN_NAME,
        'completed_epochs': completed_epochs,
        'local_run_dir': str(LOCAL_RUN_DIR),
        'drive_run_dir': str(DRIVE_RUN_DIR),
        'last_checkpoint': str(last_checkpoint),
        'best_checkpoint': str(best_checkpoint),
        'data_yaml': TRAIN_DATA_YAML,
        'imgsz': TRAIN_IMGSZ,
        'batch': TRAIN_BATCH,
        'chunk_epochs': TRAIN_CHUNK_EPOCHS,
        'total_epochs': TRAIN_TOTAL_EPOCHS,
        'updated_at_unix': time.time(),
    }
    write_drive_state(payload)
    print(f'Backed up run after epoch {completed_epochs} -> {DRIVE_BACKUP_ROOT}')


mount_drive_if_needed()
if not restore_thermal_dataset_from_drive(THERMAL_ROOT, DRIVE_THERMAL_BACKUP_ROOT, DRIVE_THERMAL_ZIP_PATH):
    raise RuntimeError(f'Thermal dataset is missing locally and no Drive backup was found at {DRIVE_THERMAL_BACKUP_ROOT} or {DRIVE_THERMAL_ZIP_PATH}')
write_dataset_yaml(THERMAL_ROOT, THERMAL_DATA_YAML, CLASS_NAMES)

state = restore_from_drive_if_needed()
completed_epochs = int(state.get('completed_epochs', 0))

print('Starting/resuming chunked training')
print('local run dir =', LOCAL_RUN_DIR)
print('drive backup dir =', DRIVE_BACKUP_ROOT)
print('completed epochs already saved =', completed_epochs)

while completed_epochs < TRAIN_TOTAL_EPOCHS:
    target_epochs = min(completed_epochs + TRAIN_CHUNK_EPOCHS, TRAIN_TOTAL_EPOCHS)
    print(f'\nTraining chunk: {completed_epochs} -> {target_epochs}')

    last_checkpoint = LOCAL_RUN_DIR / 'weights' / 'last.pt'
    if completed_epochs == 0 or not last_checkpoint.exists():
        YOLO(MODEL_PATH).train(
            data=TRAIN_DATA_YAML,
            epochs=target_epochs,
            batch=TRAIN_BATCH,
            imgsz=TRAIN_IMGSZ,
            plots=True,
            project=str(RUNS_DETECT_ROOT),
            name=RUN_NAME,
            exist_ok=True,
            save_period=1,
        )
    else:
        YOLO(str(last_checkpoint)).train(resume=True, epochs=target_epochs)

    completed_epochs = target_epochs
    backup_run_to_drive(completed_epochs)

print('\nChunked training complete.')
print('Resume checkpoint =', LOCAL_RUN_DIR / 'weights' / 'last.pt')
print('Best checkpoint =', LOCAL_RUN_DIR / 'weights' / 'best.pt')
```

```python
import os
import zipfile
from pathlib import Path

import pandas as pd

run_dir = LOCAL_RUN_DIR if 'LOCAL_RUN_DIR' in globals() else Path('/content/runs/detect/multiclass_whitehot_chunked')
if not run_dir.exists():
    raise RuntimeError(f'No training run found at {run_dir}')

print('Run dir =', run_dir)
print('Expected classes =', CLASS_NAMES)
print('Local thermal dataset =', THERMAL_ROOT)
if 'DRIVE_THERMAL_BACKUP_ROOT' in globals():
    print('Drive thermal backup =', DRIVE_THERMAL_BACKUP_ROOT)
if 'DRIVE_THERMAL_ZIP_PATH' in globals():
    print('Drive thermal zip =', DRIVE_THERMAL_ZIP_PATH)

if 'DRIVE_STATE_PATH' in globals() and DRIVE_STATE_PATH.exists():
    print('Drive state =')
    print(DRIVE_STATE_PATH.read_text(encoding='utf-8'))

results_csv = run_dir / 'results.csv'
if results_csv.exists():
    display(pd.read_csv(results_csv).tail(5))
else:
    print('results.csv not found in', run_dir)

last_model_path = run_dir / 'weights' / 'last.pt'
best_model_path = run_dir / 'weights' / 'best.pt'
print('Last checkpoint =', last_model_path)
print('Best checkpoint =', best_model_path)

if best_model_path.exists():
    trained_model = YOLO(str(best_model_path))
    print('Model class names =', trained_model.names)
else:
    print('best.pt not found in', run_dir)

zip_output_path = CONTENT_ROOT / 'runs_yolo11n_multiclass_whitehot_chunked.zip'
with zipfile.ZipFile(zip_output_path, 'w') as zip_file:
    for root, dirs, files in os.walk(run_dir):
        for file in files:
            file_path = Path(root) / file
            zip_file.write(file_path, arcname=str(file_path.relative_to('/content')))

print('Local export zip =', zip_output_path)
if 'DRIVE_BACKUP_ROOT' in globals():
    shutil.copy2(zip_output_path, DRIVE_BACKUP_ROOT / zip_output_path.name)
    print('Copied export zip to Drive =', DRIVE_BACKUP_ROOT / zip_output_path.name)
```
